## Linear Regression with One Variable (Regressão Linear Univariada)
- Model: $f_{w, b}(x^{(i)}) = w.x^{(i)} + b$

**[EN-US]**

To train a linear regression model, you want to find the best $(w,b)$ parameters that fit your dataset.  
- To compare how one choice of $(w,b)$ is better or worse than another choice, you can evaluate it with a cost function $J(w,b)$.
  - $J$ is a function of $(w,b)$. That is, the value of the cost $J(w,b)$ depends on the value of $(w,b)$.
- The choice of $(w,b)$ that fits your data the best is the one that has the smallest cost $J(w,b)$.

To find the values $(w,b)$ that gets the smallest possible cost $J(w,b)$, you can use a method called **gradient descent**. 
- With each step of gradient descent, your parameters $(w,b)$ come closer to the optimal values that will achieve the lowest cost $J(w,b)$.

The trained linear regression model can then take the input feature $x$ and output a prediction $\hat y$ = $f_{w,b}(x)$
</br></br>

**[PT-BR]**

Para treinar um modelo de regressão linear, desejamos encontrar os melhores valores para os parâmetros $(w,b)$ que se ajustem ao nosso dataset.
- Para compararseo uma escolha de $(w,b)$ é melhor ou pior que outra escolha, você pode avaliá-la com uma cost function $J(w,b)$.
  - $J$ é uma função de $(w,b)$. Ou seja, o valor do custo $J(w,b)$ depende do valor de $(w,b)$.
- A escolha de $(w,b)$ que melhor se ajusta aos seus dados é aquela que tem o menor custo $J(w,b)$.

Para encontrar os valores $(w,b)$ que obtêm o menor custo possível $J(w,b)$, você pode usar um método chamado **gradient descent**. 
- A cada posgradient descentdios, seus parâmetros $(w,b)$ se aproximam dos valores ideais que atingirão o menor custo $J(w, b)$.
- O modelo de regressão linear treinado pode então input feature entrada $x$ e gerar uma $\hat y$ = $f_{w,b}(x)$ previsão.

```python
x_train = np.array([1., 2., 3., 4.])
y_train = np.array([300., 500., 700., 900.])
print(f'X_train: {x_train}\ny_train: {y_train},\n{x_train.shape, y_train.shape}')
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = x[i] * w + b

    return f_wb
```

### Cost Function for Linear Regression with One Variable (Função de Custo para Regressão Linear com Uma Feature)
Mean Squared Error (MSE): $$J(w, b) = \frac{1}{2m} \sum\limits_{i=0}^{m-1} (f_{w, b}(x^{(i)}) - y^{(i)})^2 \\ \hat y^{(i)} = f_{w, b}(x^{(i)}) = w.x^{(i)} + b$$

**[EN-US]**
* $loss = (f_{w, b}(x^{(i)}) - y^{(i)})^2$: loss of a single example and the cost $J(w, b)$ is the sum of loss of all examples over 2 times $m$. 

**[PT-BR]**
* $loss = (f_{w, b}(x^{(i)}) - y^{(i)})^2$: loss é a perda de um único exemplo e o cost $J(w, b)$ é a soma da loss de todos os exemplo sobre 2 vezes $m$.

```python
def compute_cost_function(x, y, w, b):
    m = x.shape[0]

    cost = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost += (f_wb - y[i]) ** 2
    cost_total = cost / (2 * m)

    return cost_total
```

### Gradient Descent for Linear Regression with One Variable (Gradient Descent para Regressão Linear com uma Feature)
$$\begin{align*}
& \text{repeat until convergence (repita até convergir): for } j := 0, ..., n - 1\{ \\
& & w_j := w_j - \alpha \frac{\partial J(w, b)}{\partial w_j} \\
& & b := b - \alpha \frac{\partial J(w, b)}{\partial b} \\
& \}
\end{align*}$$

**[EN-US]**

Where, $n$ is the number of features

**[PT-BR]**

Onde $n$ é a quantidade de features

$$\frac{\partial J(w, b)}{\partial w_j} = \frac{1}{m} \sum\limits_{i=0}^{m-1} (f_{w, b}(X^{(i)}) - y^{(i)})x_j^{(i)}$$
$$\frac{\partial J(w, b)}{\partial b} = \frac{1}{m} \sum\limits_{i=0}^{m-1} (f_{w, b}(X^{(i)}) - y^{(i)})$$

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    dj_dw = dj_db = 0

    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw += (f_wb - y[i]) * x[i]
        dj_db += f_wb - y[i]
    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function):
    J_history = []
    p_history = []
    b = b_in
    w = w_in

    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(x, y, w, b)

        b -= alpha * dj_db
        w -= alpha * dj_dw

        if i < 100000:
            J_history.append(cost_function(x, y, w, b))
            p_history.append([w, b])

        if i % math.ceil(num_iters / 10) == 0:
            print(f'Iteration {i:4}: Cost {J_history[-1]:0.2e}\n', 
                  f'dj_dw: {dj_dw:0.3e}, dj_db: {dj_db:0.3e}\n', 
                  f'w: {w:0.3e}, b: {b:0.5e}\n\n')

    return w, b, J_history, p_history
```

```python
w = 0
b = 0

iterations = 10000
alpha = 1.0e-2

w, b, J, p = gradient_descent(x_train, y_train, w, b, alpha, iterations,
                              compute_cost_function, compute_gradient)
print(f'{w:8.4f}, {b:8.4f}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4))
ax1.plot(J[:100])
ax2.plot(1000 + np.arange(len(J[1000:])), J[1000:])
ax1.set_title('Cost vs. iteration (start)'); ax2.set_title('Cost vs. Iteration (end)')
ax1.set_ylabel('Cost'); ax2.set_ylabel('Cost')
ax1.set_xlabel('Iteration step'); ax2.set_xlabel('Iteration step')
plt.show
```

### MSE (Mean Squared Error)
$$J(w, b) = \frac{1}{2m} \sum\limits_{i=0}^{m-1} (f_{w ,b}(x^{(i)}) - y^{(i)})^2$$

```python
def mse(y, yhat):
    m = len(y)
    err = .0
    for i in range(m):
        err += (yhat[i] - y[i]) ** 2
    err /= (2 * m)

    return err
```

```python
y = np.array([1, 2, 3])
yhat = np.array([2, 3, 4])
mse(y, yhat)
```

## Multiple Linear Regression (Regressão Linear Múltipla)

```python
w = np.array([1., 2., 3.])
b = 4
x = np.array([10, 20, 30])
y = np.array([200, 300, 400])
```

```python
def multiple_linear_regression_non_vectorized(w, b, x):
    """
    [EN-US] Single predict using linear regression
    [PT-BR] previsão única utilizando regressão linear múltipla sem vetorização
    
    Args:
      x (ndarray): Shape (n,) example with multiple features
      w (ndarray): Shape (n,) model parameters    
      b (scalar):  model parameter     
      
    Returns:
      f_wb_x (scalar):  prediction
    """
    n = len(x)
    f_wb_x = 0

    for i in range(n):
        dot_product = w[i] * x[i]
        f_wb_x += dot_product
    f_wb_x += b

    return f_wb_x
```

```python
def multiple_linear_regression_vectorized(w, b, x):
    f_wb_x = np.dot(w, x) + b
    return f_wb_x
```

### Cost Function with Multiple Variable (Cost Function com Múltiplas Features)
$$J(w, b) = \frac{1}{2m} \sum\limits_{i=0}^{m-1} (f_{w, b}(X^{(i)}) - y^{(i)})^2$$

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
X_train.shape
```

```python
def compute_cost(X, y, w, b):
    m, _ = X.shape
    cost = 0.

    for i in range(m):
        f_wb_i = np.dot(X[i], w) + b
        cost += (f_wb_i - y[i]) ** 2
    cost /= (2 * m)

    return cost
```

### Gradient Descent with Multiple Variable (Gradient Descent com Múltiplas Features)
$$\begin{align*}
& \text{repeat until convergence (repita até convergir): for } j := 0, ..., n - 1\{ \\
& & w_j := w_j - \alpha \frac{\partial J(w, b)}{\partial w_j} \\
& & b := b - \alpha \frac{\partial J(w, b)}{\partial b} \\
&\}
\end{align*}$$

**[EN-US]**

Where, $n$ is the number of features

**[PT-BR]**

Onde $n$ é a quantidade de features

$$\frac{\partial J(w, b)}{\partial w_j} = \frac{1}{m} \sum\limits_{i=0}^{m-1} (f_{w, b}(X^{(i)}) - y^{(i)})x_j^{(i)}$$
$$\frac{\partial J(w, b)}{\partial b} = \frac{1}{m} \sum\limits_{i=0}^{m-1} (f_{w, b}(X^{(i)}) - y^{(i)})$$

```python
def compute_gradient(X, y, w, b):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):
        err = (np.dot(X[i], w) + b) - y[i]
        for j in range(n):
            dj_dw[j] += err * X[i, j]
        dj_db += err
    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db
```

```python
w = np.array([0, 0, 0, 0])
b = 0

dj_dw, dj_db = compute_gradient(X_train, y_train, w, b)
dj_dw, dj_db
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters):
    J_history = []
    w = copy.deepcopy(w_in)
    b = b_in

    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(X, y, w, b)

        w = w - alpha * dj_dw
        b -= alpha * dj_db

        if i < 100000:
            J_history.append(cost_function(X, y, w, b))

        if i % math.ceil(num_iters / 10) == 0:
            print(f'Iteration {i:4d}: Cost {J_history[-1]:8.2f}')

    return w, b, J_history
```

```python
w = np.zeros_like(w)
b = 0.

iterations = 1000
alpha = 5.0e-7

w, b, J = gradient_descent(X_train, y_train, w, b,
                          compute_cost, compute_gradient,
                          alpha, iterations)
print(f'\nw, b found by gradient descent (w, b encontrados pelo gradient descent): w = {w}, b = {b}')

m,_ = X_train.shape
for i in range(m):
    print(f'\nprediction: {np.dot(w, X_train[i]) + b:0.2f}, target value: {y_train[i]}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4))
ax1.plot(J)
ax2.plot(100 + np.arange(len(J[100:])), J[100:])
ax1.set_title('Cost vs. iteration'); ax2.set_title('Cost vs. iteration (tail)')
ax1.set_ylabel('Cost'); ax2.set_ylabel('Cost')
ax1.set_xlabel('Iteration step'); ax2.set_xlabel('Iteration step')
plt.show()
```

```python
def predict(X, w, b):
    m, n = X.shape
    yhat = np.zeros(m)

    for i in range(m):
        yhat = np.dot(X[i], w) + b
    return yhat
```

## Regularized Linear Regression (Regressão Linear Regularizada)
### Cost Function

$$J(\mathbf{w},b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})^2  + \frac{\lambda}{2m}  \sum_{j=0}^{n-1} w_j^2$$ 

```python
def cost_linear_reg(X, y, w, b, lambda_=1):
    m, _ = X.shape
    n = len(w)
    cost = .0

    for i in range(m):
        f_wb_i = np.dot(w, X[i]) + b
        cost += (f_wb_i - y[i]) ** 2
    cost /= (2 * m)

    reg = .0
    for j in range(n):
        reg += w[j] ** 2
    reg *= (lambda_ / (2 * m))

    total_cost = cost + reg
    return total_cost
```

```python
np.random.seed(1)
X = np.random.rand(5,6)
y = np.array([0,1,0,1,0])
w = np.random.rand(X.shape[1]).reshape(-1,)-0.5
b = 0.5
lambda_ = 0.7
cost = cost_linear_reg(X, y, w, b, lambda_)

print("Regularized cost:", cost)
```

### Gradient Descent

$$\begin{align*}
&\text{repeat until convergence (repita até convergir): for } j := 0, ..., n - 1 \; \lbrace \\
&  \; \; \;w_j := w_j -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial w_j}  \; \\ 
&  \; \; \;b := b -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b} \\
&\rbrace
\end{align*}$$

$$\begin{align*}
\frac{\partial J(\mathbf{w},b)}{\partial w_j}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})x_{j}^{(i)}  +  \frac{\lambda}{m} w_j \\
\frac{\partial J(\mathbf{w},b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}) 
\end{align*}$$

```python
def gradient_linear_reg(X, y, w, b, lambda_=1):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = .0

    for i in range(m):
        err = (np.dot(w, X[i]) + b) - y[i]
        for j in range(n):
            dj_dw[j] += err * X[i, j]
        dj_db += err
    dj_dw /= m
    dj_db /= m

    for j in range(n):
        dj_dw[j] += (lambda_ / m) * w[j]

    return dj_dw, dj_db
```

```python
np.random.seed(1)
X = np.random.rand(5,3)
y = np.array([0,1,0,1,0])
w = np.random.rand(X.shape[1])
b = 0.5
lambda_ = 0.7
dj_dw, dj_db =  gradient_linear_reg(X, y, w, b, lambda_)

print(f"dj_db: {dj_db}", )
print(f"Regularized dj_dw:\n {dj_dw.tolist()}", )
```
